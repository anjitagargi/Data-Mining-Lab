**HISTOGRAM PLOTTING**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

**SCATTERPLOT PLOTTING**

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generating sample data
np.random.seed(42)

# Scatter plot
x = np.random.randn(100)
y = 2 * x + np.random.randn(100)
sns.scatterplot(x=x, y=y, color='green', alpha=0.7)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### BIIVARIATE
(A) scatterPlot and the parameter is ApplicantIncomr, LoanAmount

```python
import random
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv("train.csv")

data1 = data.ApplicantIncome
data2 = data.LoanAmount

print(data1)
print(data2)

plt.scatter(data1,data2, color='skyblue', edgecolor='black')
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.title('Scatter Plot')

plt.show()
```
