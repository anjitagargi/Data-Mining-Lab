```python
import numpy as np
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate random dataset
np.random.seed(42)
random_data = np.random.randint(1, 100, 100)

# Calculate mean, mode, median, standard deviation, and variance
mean_value = np.mean(random_data)
mode_value = statistics.mode(random_data)
median_value = np.median(random_data)
std_deviation = np.std(random_data)
variance_value = np.var(random_data)

# Display the calculated values
print(f"Mean: {mean_value}")
print(f"Mode: {mode_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_deviation}")
print(f"Variance: {variance_value}")

# Create a box plot for the random dataset
plt.figure(figsize=(8, 6))
sns.boxplot(random_data)
plt.title("Box Plot of Random Dataset")
plt.show()

# Load Titanic dataset
titanic_data = pd.read_csv("titanic.csv")

# Check column names in the Titanic dataset
print(titanic_data.columns)

# Drop rows with missing age values
titanic_data = titanic_data.dropna(subset=['Age'])  # Use the correct column name 'Age'

# Create a box plot for the Titanic dataset based on age
plt.figure(figsize=(10, 8))
sns.boxplot(x='Age', data=titanic_data)
plt.title("Box Plot of Titanic Dataset based on Age")
plt.show()
```
