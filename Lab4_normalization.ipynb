### **Normalization**

normalization refers to the process of transforming numerical data into a standard scale or range to ensure that different features contribute equally to analyses and modeling. The goal is to eliminate the impact of differing units or scales in the data, making comparisons and computations more meaningful. Normalization helps in achieving better performance and stability in various data mining algorithms, particularly those sensitive to the scale of input features.

# **Example:**

Consider a dataset with two features, "Income" and "Age." The "Income" feature ranges from $20,000 to $100,000, while the "Age" feature ranges from 20 to 60 years. Since these features have different scales, a normalization technique, such as Min-Max Scaling, can be applied to bring both features within a common range, say [0, 1]. After normalization, the "Income" and "Age" values will be transformed to a standard scale, allowing algorithms to treat them uniformly without being biased towards the larger scale of income. This normalization enhances the effectiveness of data mining algorithms that rely on numerical inputs, like distance-based methods or neural networks.

### **1) Z SCORE NORMALIZATION**

Z-score standardization transforms data to have a mean of 0 and a standard deviation of 1. It's useful when the features in a dataset have different


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://github.com/shrikant-temburwar/Loan-Prediction-Dataset/raw/master/train.csv')

numerical_column = 'ApplicantIncome'

plt.figure(figsize=(6, 6))
plt.title('Distribution before Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.xlabel(numerical_column)
plt.show()

def z_score_normalization(column):
    mean = data[column].mean()
    std_dev = data[column].std()
    data[column] = (data[column] - mean) / std_dev

z_score_normalization(numerical_column)

plt.figure(figsize=(6, 6))
plt.title('Distribution after Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.xlabel(numerical_column)
plt.show()
```

### **MIN MAX NORMALIZATION**

Min-Max scaling scales the data to a specific range, often [0, 1]. It's beneficial when the features have different ranges and ensures all features contribute equally.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('https://github.com/shrikant-temburwar/Loan-Prediction-Dataset/raw/master/train.csv')

numerical_column = 'ApplicantIncome'

plt.figure(figsize=(8, 4))
plt.title('Distribution before Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

def min_max_normalization(column):
    min_val = data[column].min()
    max_val = data[column].max()
    data[column] = (data[column] - min_val) / (max_val - min_val)

min_max_normalization(numerical_column)

plt.figure(figsize=(8, 4))
plt.title('Distribution after Min-Max Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

```

### **MEAN NORMALIZATION**

Mean normalization centers the data around its mean, resulting in a distribution with a mean of 0. It helps when the dataset does not follow a normal distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://github.com/shrikant-temburwar/Loan-Prediction-Dataset/raw/master/train.csv')


numerical_column = 'ApplicantIncome'

plt.figure(figsize=(8, 4))
plt.title('Distribution before Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()


def mean_normalization(column):
    mean_val = data[column].mean()
    min_val = data[column].min()
    max_val = data[column].max()
    data[column] = (data[column] - mean_val) / (max_val - min_val)

mean_normalization(numerical_column)


plt.figure(figsize=(8, 4))
plt.title('Distribution after Mean Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

```

### **4)MAX ABSOLUTE SCALING**

Max Absolute scaling scales the data by dividing each value by the maximum absolute value in the dataset. It preserves the sign of the data and is useful when outliers are present.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://github.com/shrikant-temburwar/Loan-Prediction-Dataset/raw/master/train.csv')

numerical_column = 'ApplicantIncome'

plt.figure(figsize=(8, 4))
plt.title('Distribution before Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

def max_absolute_scaling(column):
    max_val = data[column].max()
    data[column] = data[column] / abs(max_val)

max_absolute_scaling(numerical_column)

plt.figure(figsize=(8, 4))
plt.title('Distribution after Max Absolute Scaling Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

```

### **5)Robust Scaling**

Robust scaling uses the median and interquartile range (IQR) to scale data, making it robust to outliers. It's suitable when the dataset contains outliers that might affect other scaling techniques.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('https://github.com/shrikant-temburwar/Loan-Prediction-Dataset/raw/master/train.csv')

numerical_column = 'ApplicantIncome'

plt.figure(figsize=(8, 4))
plt.title('Distribution before Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()

def robust_scaling_normalization(column):
    median_val = data[column].median()
    iqr = data[column].quantile(0.75) - data[column].quantile(0.25)
    data[column] = (data[column] - median_val) / iqr

robust_scaling_normalization(numerical_column)

plt.figure(figsize=(8, 4))
plt.title('Distribution after Robust Scaling Normalization')
sns.histplot(data[numerical_column], kde=True)
plt.show()


```
